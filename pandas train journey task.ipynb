```python
import pandas as pd
import os
import numpy as np
from datetime import datetime, date

data_loc = '../input/pandas-data-exercise-train-journeys/2019-06-03_exercise_ext/data'
files = sorted(os.listdir(data_loc))
#os.listdir(path) -> gets list of file names in specified location
#sorted() to sort files by date, otherwise they are joined in a different order

main_df = pd.read_csv('../input/pandas-data-exercise-train-journeys/2019-06-03_exercise_ext/data/2015-08-01.csv')
main_df['date']='2015-08-01'

dates = [] #useful in for loop later when getting largest values
for file in files:
    if file != 'measurement-sections.csv' and file != '2015-08-01.csv':
        n_date = file.replace('.csv', '') #get date from file name
        dates.append(n_date)
    
        temp = pd.read_csv(data_loc + f'/{file}')
        temp['date']=n_date #add date column with same date values for that file data
        main_df = pd.concat([main_df, temp])
        #print(main_df.shape) -> check to see files are being joined
        

main_df['date'] = pd.to_datetime(main_df['date']) #convert to datetime object for sorting -> with strings, it's not sorted properly
main_df #checks head and tail
```

```python
### important note - some files contain journeys made on the following day so the date may not match up (e.g. in file 2015-08-21.csv,
### some journeys have a date of 22/08 despite the file being for the 21st)

#table showing journeys - cols: date, origin_station, destination_station and count
table = main_df[['date', 'origin_station', 'destination_station']] #reduce initial table to only date/origin/dest.
new_t = table.pivot_table(columns=['date','origin_station','destination_station'], aggfunc='size').reset_index().rename(columns={0:'count'})
new_t

#pivot_table() -> can count duplicates and show them in a spreadsheet-style table
#'size' in aggfunc gives frequency
#pivot_table results in series - reset index to get dataframe, then rename last column to 'count'
#changing to dataframe using to_frame() causes an issue: column '0' is added but not in line with other columns
```

```python
#get 16 top journeys per date

top_16 = new_t[new_t['date'] == '2015-08-01'].nlargest(16, 'count')

for date in dates: #loop through dates -> create temporary df of 16 largest counts per date -> join to top_16 df
    temp = new_t[new_t['date'] == date].nlargest(16, 'count')
    top_16 = pd.concat([top_16, temp])
    
top_16.head(36)
```
