```python
import numpy as np
import matplotlib.pyplot as plt

#all plotting commands apply to the current axes
#gca() returns current axes
#gcf() returns current figure


def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

#subplot() numrows, numcols, fignum -> subplot(211) = subplot(2, 1, 1)
#commas in subplot are optional if numrows*numcols<10
#subplot() defines the graph to be plotted on in the figure as there may be multiple graphs -> 211 = 2 rows, 1 column, first graph | 212 = 2 rows, 1 column, second graph
#to manually set axes, use axes() -> axes([left, bottom, width, height]) -> all values in fractional (0 to 1) coords.
#axes() =/= axis()

plt.figure(1) #first figure -~> both graphs shown on one figure
plt.subplot(211) 
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])          # assumes 1-3 are y values, x values become 0-2
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])
#figure 1 shown

plt.figure(2)                # a second figure - shown seperately to figure 1
plt.plot([4, 5, 6])          # creates a subplot(111) by default
#figure 2 shown

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') # subplot 211 title
#figure 1 edited so shown after figure 2

#creating figurs uses memory - to release that memory, figures need to be closed with close()
```
