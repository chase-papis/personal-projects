```python
import pandas as pd
import folium
from folium import plugins
from folium.plugins import HeatMap
import geopandas as gpd
import branca.colormap as cm
from collections import defaultdict

print('imported')

map1 = folium.Map(location = [51.583332215795636, -0.0069618313646245345],
                 zoom_start = 16,
                 zoom_control = True)

#folium tile layer
tiles = ['OpenStreetMap',
         'stamenwatercolor',
         'stamentoner',
         'stamenterrain',
         'cartodbdark_matter',
         'cartodbpositron']

for tile in tiles:
    folium.TileLayer(tile, attr=tile).add_to(map1)

#custom data
data = pd.DataFrame({'house_num': [60, 87, 89, 101, 104, 120, 143, 147],
                    'lat': [51.58446, 51.58443, 51.58439, 51.58413, 51.58324, 51.58287, 51.58308, 51.58299],
                    'lon': [-0.00787, -0.00751, -0.00750, -0.00735, -0.00743, -0.00702, -0.00673, -0.00660],
                    'residents': [2, 1, 3, 4, 2, 2, 5, 7]})

heat_data = [[row['lat'], row['lon']] for index, row in data.iterrows()]
# iterrows() generates object to be able to iterate through it
# returns iterator containing the index and row information (for index, row)

color_map= cm.LinearColormap(['blue','cyan','green','yellow','red'],
                            vmin=0,
                            vmax=8).to_step(n=8)
color_map.caption ='Resident Density'
color_map.add_to(map1)

HeatMap(heat_data, 
        radius=20, 
        name='Resident Density Heatmap',
       ).add_to(map1) 
#plots it on the map

folium.LayerControl().add_to(map1)

add_rd = '''<h6>Addison Road Bus Stop</h6>
<table border=1>
<tr>
<th>Bus</th>
<th>First Bus</th>
<th>Last Bus</th>
<th>Bus Frequency</th>
</tr>
<tr>
<td>W12</td>
<td>05:20</td>
<td>00:10</td>
<td>Every 30 mins</td>
</tr>
<tr>
<td>W16</td>
<td>05:45</td>
<td>00:16</td>
<td>Every 12 mins</td>
</table>'''

popup = folium.Popup(html=add_rd, max_width=250)
folium.Marker([51.58336631319133, -0.007459966693185712],
             popup=popup).add_to(map1)

map1
```
